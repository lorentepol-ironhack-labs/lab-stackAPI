# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
import pandas as pd
import requests
import json
import time
import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
date1 = '07/06/2020'
date2 = '07/07/2020'
ts1 = round(time.mktime(time.strptime(date1, "%d/%m/%Y")))
ts2 = round(time.mktime(time.strptime(date2, "%d/%m/%Y")))
```

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate = ts1, todate = ts2)

answers = SITE.fetch('answers', fromdate = ts1, todate = ts2)
questions = questions['items']
answers = answers['items']
```

```python
answers['items']
```

```python
questions_df = pd.json_normalize(questions)
answers_df = pd.json_normalize(answers)

answers_df
```
